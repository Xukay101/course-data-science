### **Datos meteorologicos en Madrid**


---


Este conjunto de datos contiene datos meteorológicos históricos sobre madrid en un pediodo desde 1995 hasta 2015. Los datos incluyen información sobre las precipitaciones diarias, la temperatura y la velocidad del viento. Este conjunto de datos sería útil para cualquier persona interesada en estudiar el clima local o los patrones meteorológicos.

Contexto comercial: Si sabemos que el tiempo será caluroso y húmedo, se pueden tomar medidas para proteger los productos del calor y la humedad. También se pueden tomar decisiones sobre el transporte de productos sensibles al clima, como frutas y verduras.

Problema comercial: Los datos meteorológicos pueden ser imprecisos, especialmente si se trata de pronósticos a largo plazo. También pueden haber errores en los datos recopilados, lo que puede llevar a decisiones comerciales erradas.

El contexto analítico es la predicción del tiempo a partir de datos meteorológicos.

```python
# Convercion de las fechas a Datetime
df_data.CET = pd.to_datetime(df_data.CET)
df_data['Month'] = df_data['CET'].dt.month 
df_data['Year'] = df_data['CET'].dt.year
df_data['Day'] = df_data['CET'].dt.day
df_data['DayxYear'] = df_data['CET'].dt.strftime('%m %d')
```

#### Modelo para predecir la temperatura a partir de variables.

```python
# Sequential Forward Selection(sfs)
sfs = SFS(LinearRegression(),
          k_features=5,
          forward=True,
          floating=False,
          scoring = 'r2',
          cv = 0)
```

```python
X = df_data.drop(columns=[
    'Max TemperatureC', 'Min TemperatureC', 'Mean TemperatureC', 
    'CET', 'Month', 'Year', 'Day', 'Max Gust SpeedKm/h',
    'CloudCover', 'Events', 'Max VisibilityKm',
    'Min VisibilityKm', 'Mean VisibilityKm', 'DayxYear'])
y = df_data['Mean TemperatureC']
```

```python
sfs.fit(X, y)
sfs.k_feature_names_  # Lista final de features
```

```python
X = X[list(sfs.k_feature_names_)]
```

```python
# Dividir los datos en un conjunto de entrenamiento y un conjunto para test
X_train, X_test , y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
model = MLPRegressor()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

##### Evaluacion del Modelo

```python
# Metricas
print("MAE", mean_absolute_error(y_test,y_pred))
print("R2", r2_score(y_test,y_pred))
```

```python
df_temp = pd.DataFrame({'Real': y_test, 'Predicted': y_pred})
df_temp.head()
```

```python
plt.figure(figsize=(7, 8))
ax1 = sns.kdeplot(y_test, color="green", label="Real")
sns.kdeplot(y_pred, color="r", label="Predict" , ax=ax1)
plt.title('Real and prediction')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
```

```python
print(cross_validate(model, X_train, y_train, scoring='r2', return_train_score=False, cv=5))
```

```python
plt.figure(figsize=(7, 5))
plt.plot(model.loss_curve_)
plt.title("Loss Curve", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

#### Modelo para generar predicciones de la temperatura en el tiempo.

```python
tss = TimeSeriesSplit(n_splits = 3)
```

```python
# df_temperature = df_data[['CET', 'Mean TemperatureC']].set_index('CET')
df_temperature = df_data.set_index('CET')

X = df_temperature.drop(columns=[
    'Max TemperatureC', 'Min TemperatureC', 'Mean TemperatureC', 
    'Month', 'Year', 'Day', 'Max Gust SpeedKm/h',
    'CloudCover', 'Events', 'Max VisibilityKm',
    'Min VisibilityKm', 'Mean VisibilityKm', 'DayxYear'])
y = df_temperature['Mean TemperatureC']
```

```python
for train_index, test_index in tss.split(X):
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
y_train.plot(ax=ax, label='train')
y_test.plot(ax=ax, label='test')
ax.legend()
```

```python
forecaster = ForecasterAutoreg(
                regressor = DecisionTreeRegressor(random_state=42),
                lags = 12
             )

forecaster.fit(y_train)
forecaster
```

```python
y_pred = forecaster.predict(steps=y_test.shape[0])
y_pred.index = y_test.index
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
y_train.plot(ax=ax, label='train')
y_test.plot(ax=ax, label='test')
y_pred.plot(ax=ax, label='predict')
ax.legend()
```

```python
fig, ax=plt.subplots(figsize=(9, 4))
y_test.plot(ax=ax, label='test')
y_pred.plot(ax=ax, label='predict')
ax.legend()
```

```python
# Metricas
print('MAE', mean_absolute_error(y_test, y_pred))
print("R2", r2_score(y_test,y_pred))
```
